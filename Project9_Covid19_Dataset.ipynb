```python
# Dropping the UID, Last_Update, FIPS and Country_Region as there is no effect with these features on the dataset as the count of thsese categorical nature columns is 1 only i.e. the value is same of these column in each row.

df_covid = df_covid.drop(['Last_Update','Country_Region','UID','FIPS'], axis=1)
df_covid.head()
```

Thus, the above 3 columns have been dropped.

```python
# FIlling the Missing Values.

# Now let's fill the null values with the mean of the respective columns.

df_covid['Lat'].replace(np.NaN,df_covid['Lat'].mean(),inplace=True)
df_covid['Long_'].replace(np.NaN,df_covid['Long_'].mean(),inplace=True)
df_covid['Recovered'].replace(np.NaN,df_covid['Recovered'].mean(),inplace=True)
df_covid['Incident_Rate'].replace(np.NaN,df_covid['Incident_Rate'].mean(),inplace=True)
df_covid['People_Tested'].replace(np.NaN,df_covid['People_Tested'].mean(),inplace=True)
df_covid['People_Hospitalized'].replace(np.NaN,df_covid['People_Hospitalized'].mean(),inplace=True)
df_covid['Mortality_Rate'].replace(np.NaN,df_covid['Mortality_Rate'].mean(),inplace=True)
df_covid['Testing_Rate'].replace(np.NaN,df_covid['Testing_Rate'].mean(),inplace=True)
df_covid['Hospitalization_Rate'].replace(np.NaN,df_covid['Hospitalization_Rate'].mean(),inplace=True)
```

```python
# Again checking the null values after filling the same.

print(df_covid.isnull().sum())
sns.heatmap(df_covid.isnull(), cmap = 'YlOrRd')
plt.show()
```

```python
# Let's now check the outliers and clean the data.

from scipy.stats import zscore

z = np.abs(zscore(df_covid))
print(z.shape)

# Removing the outliers.

df_covid = df_covid.loc[(z<3).all(axis=1)]
print(df_covid.shape)
```

By comparing shape of 'z' and 'df_covid' shape , 10 rows have been removed as an outlier.

```python
# Treating the skewness with log function.

for col in df_covid.columns:
    if df_covid.skew().loc[col]>0.55:
        df_covid[col]=np.log1p(df_covid[col])
```

```python
# Let's Separating the input and target variables with the target variable Deaths.

X_death = df_covid.drop(columns=["Deaths"])
y_death = df_covid[["Deaths"]]
```

```python
# Let's Separating the input and target variables with the target variable Recovered.

X_rcvr = df_covid.drop(columns=["Recovered"])
y_rcvr = df_covid[["Recovered"]]
```

```python
# Scaling the Input variables

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X_death)
X
```

```python
# Importing the neccesary sklearn libraries.

from sklearn.linear_model import Lasso, LinearRegression, Ridge
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
from sklearn import linear_model
from sklearn.model_selection import GridSearchCV
```

Testing the algorithms to find the best model for the datset when the target variab,e is Deaths.

```python
# Splitting the training and testing data and finding the maximum r2 score at random state.

def max_r2_score (reg, X, y):
    max_r_scr = 0
    for r_state in range(42,101):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .20, random_state = r_state)
        reg.fit(X_train, y_train)
        pred = reg.predict(X_test)
        r2_scr = r2_score(y_test, pred)
        if r2_scr > max_r_scr:
            max_r_scr = r2_scr
            final_r_state = r_state
            
    print("Maximum r2 score :", max_r_scr, "achieved at", final_r_state)
    print('\n')
    print("Mean Absolute Error : ",np.sqrt(mean_absolute_error(y_test, pred)))
    print('\n')
    print("RMSE : ",np.sqrt(mean_squared_error(y_test, pred)))
```

```python
# Finding the best alpha value for Ridge regression using GridsearchCV.

rg = Ridge()
parameters = {"alpha":[0,0.0001,0.001,0.01,0.1,1]}
gd = GridSearchCV(rg, parameters)
gd.fit(X_death,y_death)
gd.best_params_
```

```python
# Finding the best alpha value for Ridge regression using GridsearchCV.

lreg = Lasso()
parameters = {"alpha":[0,0.0001,0.001,0.01,0.1,1]}
gd = GridSearchCV(lreg, parameters)
gd.fit(X_death,y_death)
gd.best_params_
```

```python
# Creating the instances for the algorithms.

lr = LinearRegression()
lreg = Lasso(alpha=0.001)
rg = Ridge(alpha=0.1)
sv = SVR()
dtr = DecisionTreeRegressor()
knnr = KNeighborsRegressor()
rfr = RandomForestRegressor()
```

```python
models = []
models.append(('LinearRegression', lr))
models.append(('Lasso', lreg))
models.append(('Ridge', rg))
models.append(('SVR', sv))
models.append(('DecisionTreeRegressor', dtr))
models.append(('KNeighborsRegressor', knnr))
models.append(('RandomForestRegressor', rfr))
```

```python
# CReating a list which holds the algorithms.

model = []
cvs = []

for name,model in models:
    k=model
    print('--------',name,'--------')
    r_state = max_r2_score(k, X_death, y_death)
    print('\n')
    mR2 = cross_val_score(k, X_death, y_death, cv=5, scoring = "r2").mean()
    print('Cross validation score (Mean r2 score):', mR2)
    stdR2 = cross_val_score(k, X_death, y_death, cv=5, scoring = "r2").std()
    print('Standard Deviationin r2 score :', stdR2)
    print('\n')
    print('\n')
```

LinearRegression is giving the r2 score of 0.997 at random state 88. Thus, it is selcted as a final model for this dataset when the target variable or when "Deaths" is predicted.

```python
# Testing the algorithms to find the best model for the datset when the target variable is Recovered.

model = []
cvs = []

for name,model in models:
    k=model
    print('--------',name,'--------')
    r_state = max_r2_score(k, X_rcvr, y_rcvr)
    print('\n')
    mR2 = cross_val_score(k, X_rcvr, y_rcvr, cv=5, scoring = "r2").mean()
    print('Cross validation score (Mean r2 score):', mR2)
    stdR2 = cross_val_score(k, X_rcvr, y_rcvr, cv=5, scoring = "r2").std()
    print('Standard Deviationin r2 score :', stdR2)
    print('\n')
    print('\n')
```

Ridge Regression is giving the r2 score of 0.95 at random state 53. Thus, it is selcted as a final model for this dataset when the target variable or when "Recovered" is predicted.
